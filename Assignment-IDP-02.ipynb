```python
from math import pi as p
from math import sqrt as sq
from math import exp as e

sqrt_pi = sq(2 * p)

# normal pdf function: pdf stands for probability density function
def nor_pdf(x, mu=0, sigma=1):
    return (e(-(x - mu) ** 2 / 2 / sigma ** 2) / (sqrt_pi * sigma))
```

```python
xs = [x/10 for x in range(-50, 50)] # x-axis
x_pdf = [nor_pdf(x/10) for x in range(-50, 50)] # y-axis
```

```python
from matplotlib import pyplot as plt
plt.plot(xs, x_pdf, '-', label='mu=0, sigma=1')
plt.plot(xs, [nor_pdf(x, sigma=2) for x in xs],'--', label='mu=0, sigma=2')
plt.plot(xs, [nor_pdf(x, sigma=3) for x in xs],'--', label='mu=0, sigma=3')
plt.legend(loc='upper right')
plt.title("normal pdf graph")
plt.savefig("stat.pdf", dpi=1200)
```

```python
plt.plot(xs, [nor_pdf(x, mu=1) for x in xs], label='mu=1, sigma=0')
plt.plot(xs, [nor_pdf(x, mu=2) for x in xs], label='mu=2, sigma=0')
plt.plot(xs, [nor_pdf(x, mu=0) for x in xs], label='mu=-1, sigma=0')
```
